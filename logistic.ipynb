```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return (1 / (1 + np.exp(-z)))
```

```python

def gradient_descent(X, y, theta, alpha, num_iterations):

    J_all = []

    for i in range(num_iterations):
        # Compute the predictions.
    #     print(X.shape, theta.shape)
        p = sigmoid(np.dot(X, theta))

        # Compute the cost function.
        J = -np.sum(y * np.log(p) + (1 - y) * np.log(1 - p)) / X.shape[0]

        gradients = np.dot(X.T, (p - y)) / X.shape[0]

        # Update the model parameters.
        theta -= alpha * gradients
        print(theta, J)

        # Store the cost function value for this iteration.
        J_all.append(J)

    return theta, J_all


num_samples = 500
num_features = 3

X = 2 * np.random.rand(num_samples, num_features)
true_theta = np.array([0.0, -5, 0.03])
noise = 3
logits = X.dot(true_theta) + noise
probabilities = 1 / (1 + np.exp(-logits))
y = (probabilities > 0.5).astype(int)
print(y)
X = np.c_[np.ones((X.shape[0], 1)), X]

theta = np.zeros((num_features+1))
alpha = 0.01

num_iterations = 1000

theta, J_all = gradient_descent(X, y, theta, alpha, num_iterations)
print(noise, true_theta)

# print(theta)
# print(J_all)
plt.plot(range(num_iterations), J_all)
J_all[-1]
```

```python
import numpy as np

# Generate some example data
np.random.seed(0)
num_samples = 10000
num_features = 3

# Generate random feature data (X)
X = np.random.randn(num_samples, num_features)

# Generate labels (y) where 1 represents a positive class and 0 represents a negative class
true_theta = np.array([0.5, -0.5, 1.0])
logits = X.dot(true_theta)
probabilities = 1 / (1 + np.exp(-logits))
y = (probabilities > 0.5).astype(int)

# Hyperparameters
learning_rate = 0.01
num_iterations = 10000

# Initialize parameters (coefficients and bias)
theta = np.random.randn(num_features)
bias = np.random.randn()

# Gradient Descent
for iteration in range(num_iterations):
    logits = X.dot(theta) + bias
    probabilities = 1 / (1 + np.exp(-logits))
    
    # Compute the gradient of the cross-entropy loss
    gradient_theta = -X.T.dot(y - probabilities) / num_samples
    gradient_bias = -np.sum(y - probabilities) / num_samples
    
    # Update parameters using gradient descent
    theta -= learning_rate * gradient_theta
    bias -= learning_rate * gradient_bias

# The 'theta' values now represent the learned coefficients, and 'bias' represents the bias term
print("Coefficients (theta):", theta)
print("Bias (intercept):", bias)
```
